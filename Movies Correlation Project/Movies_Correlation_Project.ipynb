```python
# First let's import the packages I will use in this project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Read in the data
df = pd.read_csv('movies.csv')
```

```python
#Let's look at the data
df.head(5)
```

```python
#Let's see if there is any missing data

for col in df.columns:
    pct_missing = round(np.mean(df[col].isnull()),2)
    print('{} - {}%'.format(col, pct_missing))
    
#From the result,It has some percentage of missing value but I think it's not impact for analyzing 
```

```python
#Data type for my columns
print(df.dtypes)
```

```python
#Create correct year column
#The reason why I create new one,I found some data in 'year' column not equal with year in 'released' column

df['yearcorrect'] = df['released'].str.split(pat=',',expand=True)[1].str[:5]
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False).head()
```

```python
pd.set_option('display.max_rows', 100)
```

```python
#Drop any duplicates

df.drop_duplicates()
```

```python
#what things are most correlated to this gross revenue?

#I guess budget have the most correlation with gross
```

```python
#Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget VS Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')

plt.show()
```

```python
#Regression plot Budget VS Gross using seaborn

sns.regplot(data = df, x='budget', y='gross', scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# As we knows,budget and gross have correlation but we don't know how much.
```

```python
#Let's start looking at correlation (Numeric data)
df.corr()

#the budgets and gross have a high correlation.
```

```python
#High correlation between budget and gross
#I was right
```

```python
#Create visualization of correlation (Numeric data)

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movies Features')
plt.ylabel('Movies Features')

plt.show()
```

```python
#Look at non-numeric data
#I will transform non-numeric data to category for create correlation.


df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
#Create visualization of correlation (All data)

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movies Features')
plt.ylabel('Movies Features')

plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values(ascending=False)
sorted_pairs
```

```python
#To fecth only high correlation

high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
